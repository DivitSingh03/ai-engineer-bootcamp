```python
# Problem 1 - Reverse a string 
text = "Hello"
reversed_text = ""

for char in text:
    reversed_text = char + reversed_text

print(reversed_text)
```

```python
# Problem 2 - Check palindrom

text = input("Enter a string to check if it is a palindrome : ")
reversed_text = "".join(reversed(text))

if text == reversed_text:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# Problem 3 - Char frequency

text = input("Enter a string :")

char_count = {}

for char in text:
    if char not in char_count.keys():
        char_count[char] = 1
    else:
        char_count[char] += 1

print(char_count)
        
```

```python
# Problem 4 - Calculate Factorial

num = int(input("Enter a number to calculate its factorial : "))

fact = 1

while num != 0:
    if num > 1:
        fact = fact * num
    else:
        fact = fact * 1
    num = num - 1

print(fact)
```

```python
# Problem 6 - FizzBuzz

for i in range(1,101):
    if i % 3 == 0:
        if i % 5 == 0:
            print("FizzBuzz")
        else:
            print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```
